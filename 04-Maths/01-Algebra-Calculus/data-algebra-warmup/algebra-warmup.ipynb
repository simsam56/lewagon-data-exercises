# Plotting points

Today's first orders of business:
1. Plot a point
1. Plot multiple points

To plot (unconnected) points, we will use [`plt.scatter()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html#matplotlib.pyplot.scatter.).

It can take many arguments, but the first two positional arguments are the most important: the `x` and the `y`.

## Plot One Point

Plot a dot at **x=5** and **y=10**

```python
dot_coordinates = [5, 10]
plt.scatter(dot_coordinates[0], dot_coordinates[1])
```

## Plot Multiple Points

Plot several dots stored in a list of **dimensions (2, 13)** (meaning a list of lists).

Hint: the `x` and `y` arguments can be lists of x-es and y-s.

```python
dots = [
    [1, 1, 2, 2, 2, 3, 1, 3, 3, 3, 3, 0, 0],
    [2, 3, 2, 4, 1, 3, 4, 2, 0, 5, 7, 2, 3]
]

plt.scatter(dots[0], dots[1])
```

In this case we used lists for x and y, but this would work equally well with NumPy 1D-arrays or Pandas Series.

## Plotting a Line

Plot a line starting at **x=4, y=0** and ending at **x=20, y=4**. The function you should use is [pyplot.plot()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot).

`plt.plot()` will plot **the points** you give it, and draw **a straight line between the points**. So, to plot a line, you only need to provide two points and `plt.plot()` will draw the line between the two.

Just like `plt.scatter()`, `plt.plot()` expects two arguments: a list of x-es and a list of y-s.

```python
start = {"x": 4, "y": 0}
end = {"x": 20, "y":4}

plt.plot([start["x"], end["x"]],[start["y"],end["y"]])
```

## Plotting Functions

To plot a mathematical function, we'll always follow this recipe:

1. Define the range of x-values you want to plot
1. Create a list or array of evenly spaced x-values within that range
1. Calculate the corresponding y-values for those x-values using your mathematical function
1. Plot the x- and y-values

### Plotting One Function

‚ùì Generate data points within the range `[-10, 10]`. These points are your `x` points. Use them to plot the following function:

$$y = 4x + 2$$


Use [ plt.ylim(-10, 10)](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.ylim.html#matplotlib.pyplot.ylim) to keep the **y-axis** between **-10 and 10**

<br/>

<details>
  <summary>üí° Hint</summary>
  You can use <code><a href="https://docs.scipy.org/doc/numpy/reference/generated/numpy.arange.html">numpy.arange()</a></code> or <code><a href="https://numpy.org/doc/stable/reference/generated/numpy.linspace.html">numpy.linspace()</a></code> to generate a list of N numbers in a range.

  The difference?
  - Both take a start and stop value
  - With `np.arange()` you define the step size for the points in-between
  - With `np.linspace()` you define the number of points in-between
</details>

```python
x = range(-10,10,1)
y=[4*i+2 for i in x]

plt.plot(x,y)
plt.ylim(-10,10)
```

### Plotting More Advanced Functions

‚ùì Plot these 4 functions together in a new graph, using the same set of `x` points generated above, keeping the **y-axis** between **-2 and 6** and the **x-axis** between **-4 and 4**.  

$$y1 = x^2$$

$$y2 = e^x$$

$$y3 = \frac{1}{1 + e^{-(-3 + 2*x)}}$$

<details>
  <summary>üí° Hints</summary>
    
  There are multiple ways to do this... to name a few:
  - Use a `for` loop/list comprehension
  - Use NumPy vectorial functions (e.g. `np.exp(your_list_of_x_numbers)`
  - Use the Python built-in function [`map`](https://www.w3schools.com/python/ref_func_map.asp)

  Pro tip: use NumPy vectorial functions: write out the function like it's math. It's not only faster, but also easier to read!

  You'll need `np.exp()`.

</details>

Are you getting a very angular line? Try with a smaller step size. Remember that `plt.plot()` just plots the x-es and y-s you give it and draws a **straigh** line between them. The closer your points are to each other, the smoother your line will be.

```python
x = np.linspace(-2, 6, 100)
y1=[i**2 for i in x]
y2=[np.exp(i) for i in x]
y3=[1/(1+np.exp(-(-3+2*i))) for i in x]

plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)

plt.ylim(-2,6)
plt.xlim(-4,4)
```

## Combining points and lines

We have a **scatter plot** generated below with `sklearn`, a library you will learn soon! We want to draw a **straight red dotted line** that follows the dots' distribution as closely as possible üñåÔ∏è

Our final result should look like this:  
<img src="https://wagon-public-datasets.s3.amazonaws.com/data-science-images/02-Data-Toolkit/03-Data-Visualization/020302-regression-line-example.png" alt="regression_line" width=400>

```python
from sklearn import datasets

x, y, _ = datasets.make_regression(n_samples=100, n_features=1, noise=10, coef=True, random_state=0)

plt.scatter(x,y);
```

‚ùì No need for fancy calculations, just estimate what the line should be, using your eyes only.

Plot a dotted red line **approximately** regressing the scatter plot above   
- Reuse the `x` and `y` we generated in the previous cell
- Remember: two points are enough to draw a line with `plt.plot()`
- To make your line red and dotted use these arguments: `color='r', linestyle='--'` <br>(play around with other values for `color`: `'g'` or `'c'` or `'m'` and for `linestyle`: `':'` or `'-.'`)

```python
start = {"x": -2, "y": -100}
end = {"x": 2, "y":60}

plt.plot([start["x"], end["x"]],[start["y"],end["y"]],color='r', linestyle='-', linewidth=2)


x, y, _ = datasets.make_regression(n_samples=100, n_features=1, noise=10, coef=True, random_state=0)

plt.scatter(x,y);
```

‚ùì Now, we'll add some context to our graph by adding:
1. A title saying "Linear Regression"
1. A legend describing the data points and the line
1. Labels for the x-axis ("input value") and the y-axis ("output value")

We give you the code here. Just replace the `# YOUR CODE HERE` by the plot you created earlier, but this time add a `label` for each of the two plots.

```python
# Set the size of the figure
fig = plt.figure(figsize=(10, 5))

# Plot the scatter plot and the accompanying red line
# This time add a `label` to the two plots
plt.scatter(x, y, label="data points")
plt.plot([start["x"], end["x"]], [start["y"], end["y"]],
         color='r', linestyle='--', label="regression line")


# Add title and legend
plt.title("Linear Regression")
plt.legend()

# Add labels for both axes
plt.xlabel("input value")
plt.ylabel("output value")

# Add annotation to the regression line
plt.annotate("Regression line", (0,0), (1,-50), arrowprops={"arrowstyle":"->"});
```
